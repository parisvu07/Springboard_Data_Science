We know that tweets are given a sentiment number, and we want to break it down for better analysis. Since the author of this dataframe did not specify what each number means, we will assume that: \
5 = positive \
3 = neutral \
1 = negative

```python
#Check for unique values in 'sentiment' column
df.sentiment.unique()
```

```python
#Since the author did not specify what 5, 3, and 1 stands for, we assume 5 is positive, and 1 is negative
df = df[df['sentiment'] != 'not_relevant']
df = df[df['sentiment'] != "3"]
```

This dataframe is about Apple and its tweets, but we also want to be careful about the queries that the author used to pull this information

```python
#Check for keyword(s) in query
df['query'].unique()
```

After we've checked all of the misterious columns, we now then drop irrelevant columns and just leave "sentiment", "sentiment:confidence", "date", and "text" with the dataframe

```python
#Drop irrelevant columns
df = df.drop(['_unit_id','_golden','_unit_state','_trusted_judgments','_last_judgment_at','query','id','sentiment_gold'],axis = 1)
```

```python
df.head()
```

To make future analysis easier, we will lowercase all of the letters within the text column in the df dataframe

```python
df['text'] = df['text'].str.lower()
```

After we've dropped all of the columns that we think were irrelevant, the only sentiment values left in the dataframe are 5 (positive), and 1 (negative). To simplify the numbers, we create the function __new_sentiment(x)__ to convert 5 to 1, signifying positive, and 1 to 0, signifying negative

```python
#Double check sentiment
df.sentiment.unique()
```

```python
#Tweet Sentiment
#Positive Sentiment = 1
#Negative Sentiment = 0

def new_sentiment(x):
    if x == "5":
        return 1
    else:
        return 0
    
df.sentiment = df.sentiment.apply(new_sentiment)
```

```python
df.sentiment
```

By checking for unique values within the "product" column, we can see that this dataframe not only contain Apple tweets, but it also contains Android related tweets. Because of that, we will be dropping all the rows that do not refer to Apple products

```python
df2 = df2[df2['product'] != 'Google']
df2 = df2[df2['product'] != 'nan']
df2 = df2[df2['product'] != 'Android']
df2 = df2[df2['product'] != 'Android App']
df2 = df2[df2['product'] != 'Other Google product or service']
```

Similar to how we cleaned the df dataframe, we want to also lowercase all of the tweets in the df2 dataframe, and leave df2 to the side for later analysis

A few steps ago when we called df2.info(), we see that the "product" column has a significantly lower non-null values than the "text" and the "sentiment" column. To find the exact % of missing values, we compute the variable "missing"

```python
#count missing value from df2
missing = pd.concat([df2.isnull().sum(), 100 * df2.isnull().mean()], axis=1)
missing.columns=['count','%']
missing.sort_values(by=['%'], ascending=False)  
```

Due to the high % of missing products, we want to see whether or not the tweets associated with them mentioned any Apple products. To do this, we transformed the product to a binary column with True indicating "missing data", and False indicating "no missing data".

```python
df3 = df2.copy()
df3['product'] = df3['product'].isnull()
df3.head()
```

We continue to filter out just only rows that have missing products and assign it to the variable __df3_missing_products__. We then create another variable called __df3_apple_related__ to store rows that are missing product data, but their associated tweets mentioned the keywords iphone, ipad, or apple. 

```python
df3_missing_products = df3[df3['product'] == True]
```

```python
#Filter for text that does not contain related words
df3_apple_related = df3_missing_products[df3_missing_products['text'].str.contains('iphone|ipad|apple', na=False)]
df3_apple_related.info()
```

```python
df3_apple_related.head()
```

By calling .info() on __df3_apple_related__, we can see that all the columns in there have 3118, meaning: \

## Combining 

With df2 being the original data with lowercased tweets, we want to eliminate only rows that have both missing values in the "product" column and unrelated tweets. To do this, we will drop all N/A values from the "product" column in __df2__ dataframe and concatenate this with __df3_apple_related__ dataframe. The df3_apple_related dataframe contains rows that have missing values in "product", but Apple related tweets. We store the result in __df4__.

```python
#take out only rows that have a product
df2 = df2[df2['product'].notna()]
```

```python
df2.info()
```

```python
#df4 = df2 + df3
df4 = pd.concat([df2, df3_apple_related])
```

## Word Cloud using established datasets (df and df2)

Because the initial purpose is to create a word cloud, we will create (3) word clouds for __df__, __df4__, and __df_df4__ (df and df4 combined). To create df_df4, we merge df and df4 together on text. This method allows us to see if the data in df4 were pulled during the same period with df since it did not contain any dates.

```python
df_df4 = df.merge(df4, on='text', how='outer', suffixes=('_1', '_4'))
```

```python
df_df4.head()
```

```python
df_df4.info()
```

Looking at the uneven number of non-null values in each column, it is clear that the two dataframes (df and df4) were pulled separately and do not relate. It is totally fine, we were just exploring data and it does not hurt to give it a try!

```python
# Generate a word cloud image for df_df4
text = " ".join(review for review in df_df4.text)
wordcloud = WordCloud(background_color="white").generate(text)

# Display the generated image:
# the matplotlib way:
plt.figure(figsize=(10,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# Generate a word cloud image for df
text = " ".join(review for review in df_df4.text)
wordcloud = WordCloud(background_color="white").generate(text)

# Display the generated image:
# the matplotlib way:
plt.figure(figsize=(10,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# Generate a word cloud image for df4
text = " ".join(review for review in df_df4.text)
wordcloud = WordCloud(background_color="white").generate(text)

# Display the generated image:
# the matplotlib way:
plt.figure(figsize=(10,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

Both df and df4 have a common 'big' word of __SXSW__. After doing a simple google search, SXSW is a music festivals that are usually held in mid-March. It is surprising to see that both df (with dates in December 2014) and df4 (unknown dates) mention this festival. 

From Wikipedia: "South by Southwest, abbreviated as SXSW and colloquially referred to as South By, is an annual conglomeration of parallel film, interactive media, and music festivals and conferences organized jointly that take place in mid-March in Austin, Texas, United States."
